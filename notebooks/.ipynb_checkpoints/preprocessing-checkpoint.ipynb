# removing punctuation

## create a function

```python
import string

def remove_punctuation(text):
    for punctuation in string.punctuation:
        text_clean= text.replace(punctuation, '')
    return text_clean
```

```python
data_clean=data.copy()
for punctuation in string.punctuation:
    data_clean=data_clean.apply(lambda x: x.str.replace(punctuation,""))
data_clean.head()
```

## removing numbers

## create a function

```python
def remove_numbers(text):
    text=str(text)
    text_clean=''.join(char for char in text if not char.isdigit())
    #text_clean=text.str.replace(r'\d+', '',)
    return text_clean
```

```python
data_clean2=data_clean.copy()
```

```python
data_clean2.head()
```

```python
data_clean2['Top1']=data_clean2.Top1.apply(remove_numbers)
```

```python
for column in data_clean2.columns:
    data_clean2[column]=data_clean2[column].apply(remove_numbers)
```

# remove stopwords

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords 

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    tokens = word_tokenize(text)
    tokens_cleaned = [w for w in tokens if not w in stop_words]
    for w in tokens_cleaned:
        text_clean=' '.join(tokens_cleaned)
    return text_clean
```

```python
data_clean3=data_clean2.copy()
```

```python
data_clean3.head()
```

```python
for column in data_clean3.columns:
    data_clean3[column]=data_clean3[column].apply(remove_stopwords)
```

```python
data_clean3.head()
```

# lemmatize

```python
from nltk.stem import WordNetLemmatizer

def lemmatize_text(text):
    tokens = word_tokenize(text)
    verb_lemmatized = [                  
                        WordNetLemmatizer().lemmatize(word, pos = "v") 
                        for word in tokens
                    ]
    noun_lemmatized = [                 
                        WordNetLemmatizer().lemmatize(word, pos = "n")
                        for word in verb_lemmatized
                        ]
    
    text_clean=' '.join(noun_lemmatized)
    return text_clean
```

```python
data_clean4=data_clean3.copy()
```

```python
data_clean4.head()
```

```python
for column in data_clean4.columns:
    data_clean4[column]=data_clean4[column].apply(lemmatize_text)
```
